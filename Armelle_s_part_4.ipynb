```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial import legendre as L
```

```python
def degree(n):
    if n!=0:
        list_=[0]*(n+1)
        list_[-1]=1
        return list_
    if n==0:
        return [1,0] #small fix, [1,0] is equivalent to [1] but they didn't like [1]
```

```python
def norm_legendre(j): #returns normalized legendre defined on [0,1]
    return np.sqrt(2*j+1)*L.Legendre(degree(j), domain=[0,1])
```

```python
def pdf_w(x, n):
    total=0
    for j in range(0,n+1):
        total+=norm_legendre(j)(x)**2
    return total/(n+1)
```

```python
def l_2(x,i): #evaluates the square of the ith normalized legendre polynomial at the value x
    return (2*i+1)*(L.Legendre(degree(i), domain=[0,1])(x) )**2
```

- Using https://math.stackexchange.com/questions/1428311/how-to-show-that-any-legendre-polynomial-over-1-1-is-bounded-by-one-in-ab  Legendre polynomial are bounded in absolute value by 1

- We're considering the renormalized Legendre polynomials, hence the nth Legendre polynomial is bounded in absolute value by sqrt(2n+1)

- Hence the square of a Legendre polynomial is lower bounded by 0 and upper bounded by 2n+1


In order to sample from 1/w we will do rejection sampling (justify in the report why you're allowed to do that). The method can be found here: https://www.stats.ox.ac.uk/~rdavies/teaching/PartASSP/2020/lectures_latest/simulation_lecture3.pdf



```python
def gen_from_w(N, M_wanted, seed): #N the number of Legendre polynomials considered, M the number of wanted samples 
    #(as a rule of thumb, we'll have to generate 100 times more uniform r.v), M=1000 is nice
    
    np.random.seed(seed)
    
    up_bound=np.sum(2*np.arange(0,N+1)+1)/(N+1) #upper bound of 1/w
    
    #rejection sampling 
    
    M=100*M_wanted
    
    Y=np.random.uniform(low=0.0, high=1.0, size=M)
    V=np.random.uniform(low=0.0, high=up_bound, size=M)
    
    index_list=[]
    
    for m in range(M):
        if V[m]<pdf_w(Y[m], N):
            index_list.append(m)
    return Y[index_list], V[index_list] #the first output is our wanted generated random variables, the second output
# is used for the scatter plot justifying the method works well 
```

```python
N=10 #we fix some value for n
seed=2552 #for reproducibility
np.random.seed(seed)

up_bound=np.sum(2*np.arange(0,N+1)+1)/(N+1)
```

```python
M=10000 #sample size 

Y=np.random.uniform(low=0.0, high=1.0, size=M)
V=np.random.uniform(low=0.0, high=up_bound, size=M)

X_list=[]
index_list=[]

for m in range(M):
    if V[m]<pdf_w(Y[m], N):
        X_list.append(Y[m])
        index_list.append(m)

X=np.array(X_list)
```
